```python
import json
import matplotlib.pyplot as plt

with open("land.json", encoding="utf-8") as fil:
    data = json.load(fil)

data[0]

```

```python
verdensdeler = {}
for land in data:
    print(land['verdensdel'])
    if land ['verdensdel'] not in verdensdeler:
        print("fant en ny verdensdel! Putter den i ordboka")
        verdensdeler[land['verdensdel']] = 1
    else:
        verdensdeler[land['verdensdel']] += 1

print(verdensdeler)
```

```python
land = list(verdensdeler.keys()) # gjør nøklene til en liste
antall = list(verdensdeler.values()) # gjør verdiene til en liste

# Oppgave
plt.bar(land, antall)
plt.title("Antall land på topplista")
plt.xlabel("tekst på x-aksen")
plt.ylabel("tekst på y-aksen")
plt.style.use("fivethirtyeight")
```
